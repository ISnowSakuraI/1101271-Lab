**Problem 1** (1 คะแนน)

---

1. นาย C จะต้องจัดเก็บประวัติลูกค้า โดยมีข้อมูลสำคัญที่ต้องจัดเก็บ คือ ชื่อ วันเดือนปีเกิด และเบอร์โทรศัพท์

**Command 1: ** ให้จัดเก็บข้อมูลของลูกค้าโดยใช้ Array of String ในการเก็บรวบรวมข้อมูลดังนี้
*   นาย A, 11/04/1988, 0861112222
*   นาย B, 25/01/2000, 0893334422
*   นาย C, 08/11/1992, 0820400202

**Command 2: ** ให้แปลง Array เป็น DataFrame และแสดงผลข้อมูลดังกล่าวในรูปแบบตาราง

**Command 3: **ให้แก้ไขข้อมูลวันเดือนปีของ นาย B จาก 25/01/2000 เป็น 22/02/2000 และให้เพิ่มข้อมูลของนาย D ด้วย (นาย D, 04/03/2011, -)  

**Command 4: **ให้แสดงผลข้อมูลหลังแก้ไขเป็นรูปแบบตารางเหมือนคำสั่งที่ 2


Input : รายการลูกค้า

Output : ตารางแสดงผล


Hint:
*   df.at(rows,column) df._append(data, ignore_index=True), Dictionary
*   Array และ DataFrame


```python
############################################################
#                      Answer: Problem 1                   #
############################################################

import pandas as pd

# Command 1: เก็บข้อมูลใน Array
data = [
    ['นาย A', '11/04/1988', '0861112222'],
    ['นาย B', '25/01/2000', '0893334422'],
    ['นาย C', '08/11/1992', '0820400202']
]

# Commnad 2: แปลงเป็น DataFrame และแสดงผลข้อมูลเป็นตาราง (Hint: DataFrame)
columns = ['ชื่อ', 'วันเดือนปีเกิด', 'เบอร์โทรศัพท์']
df = pd.DataFrame(data, columns=columns)
df

# Commnad 3: แก้ไขข้อมูล DataFrame โดยแก้วันเดือนปีของนาย B และเพิ่มข้อมูลนาย D
df.at[1, 'วันเดือนปีเกิด'] = '22/02/2000'
new_data = {'ชื่อ': 'นาย D', 'วันเดือนปีเกิด': '04/03/2011', 'เบอร์โทรศัพท์': '-'}
df_new = pd.DataFrame([new_data])
df = pd.concat([df, df_new], ignore_index=True)

# Command 4: แสดงผล DataFrame หลังปรับปรุงข้อมูลนาย B และ D แล้ว
df

```

**Problem 4** (2 คะแนน)

---

Command 1: จงเลือกข้อมูลเฉพาะวันที่ (Day) จังหวัด (Province) และอุณหภูมิ (Temperature) ของวันที่ "1/1/2562" และเรียงลำดับตามอุณหภูมิจากมากไปน้อย

Command 2: จงออกแบบกราฟที่นำเสนอข้อมูลในตาราง โดยแสดงผลลัพธ์ดังภาพ

```python
############################################################
#                      Answer: Problem 4                   #
############################################################
import matplotlib.pyplot as plt

# Command 1: Add Code Here
df_filtered = df1.loc[
    df1['Day'] == '1/1/2562',
    ['Day', 'Province', 'Temperature']
]
df3 = df_filtered.sort_values(by='Temperature', ascending=False)
df3

# Command 2: Add Code Here
plt.bar(df3['Province'], df3['Temperature'])
plt.show()
```

**Problem 5** (2 คะแนน)

---

Command 1: จงเลือกข้อมูลเฉพาะวันที่ (Day) จังหวัด (Province) และอุณหภูมิ (Temperature) ของจังหวัด "Bangkok" ใส่ Data Frame หนึ่ง และจังหวัด "Phuket" ใส่อีก Data Frame หนึ่ง และเรียงลำดับตามวันที่จากน้อยไปมาก

Command 2: จงออกแบบกราฟที่นำเสนอข้อมูลในตารางทั้งสอง โดยแสดงผลลัพธ์ดังภาพ
image.png

```python
############################################################
#                      Answer: Problem 5                   #
############################################################

# Command 1: Add Code Here
df_bkk = df1.loc[
    df1['Province'] == 'Bangkok',
    ['Day', 'Province', 'Temperature']
].sort_values(by='Day')

df_phu = df1.loc[
    df1['Province'] == 'Phuket',
    ['Day', 'Province', 'Temperature']
].sort_values(by='Day')

df_bkk
df_phu

# Command 2: Add Code Here
plt.figure()
plt.plot(df_phu['Day'], df_phu['Temperature'],
         label='Phuket', color='green', linestyle='--')

plt.plot(df_bkk['Day'], df_bkk['Temperature'],
         label='Bangkok', color='blue', linestyle='--')

plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Temperature of Bangkok and Phuket')

plt.legend()
plt.show()
```
