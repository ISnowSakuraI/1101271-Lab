1.จงสร้างแบบจำลองเพื่อจัดกลุ่มข้อมูลของดอกไอริส ดังนี้
* สามารถนำเข้าข้อมูล wine data จาก sklearn datasets ได้
* สามารถแปลงข้อมูล wine data ให้เป็นตารางใน DataFrame พร้อมแสดงผลออกทางจอภาพ
* สามารถสร้างแบบจำลองโดยใช้เทคนิค K-means clustering โดยกำหนดให้ k = 3 พร้อมแสดงข้อมูลผลลัพธ์ (wine target/y_true) และข้อมูลที่ได้จากการทำนาย (y_pred) ออกทางจอภาพได้
* สามารถแสดงแผนภาพกราฟจุดของข้อมูล y_true และ y_pred ออกทางจอภาพ โดยแกนนอนคือข้อมูลในคอลัมน์แรก แกนตั้งคือข้อมูลในคอลัมน์ที่สอง
*   สามารถแสดง Confusion Matrix ของ y_true และ y_pred พร้อมทั้งประเมินแบบจำลองโดยใช้ Accuracy และแสดงออกทางจอภาพได้


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn import metrics

#สามารถนำเข้าข้อมูล Wine data จาก sklearn datasets ได้ถูกต้อง
wine = datasets.load_wine()

#สามารถแปลงข้อมูล Wine data ให้เป็นตารางใน DataFrame พร้อมแสดงผลออกทางจอภาพ
df = pd.DataFrame(wine.data, columns=wine.feature_names)
print(df)

#สามารถสร้างแบบจำลองโดยใช้เทคนิค K-means clustering โดยกำหนดให้ k = 3
model = KMeans(n_clusters=3, random_state=42)
model.fit(df)

#สามารถแสดงข้อมูลผลลัพธ์ (wine target/y_true) และข้อมูลที่ได้จากการทำนาย (y_pred) ออกทางจอภาพได้
y_true = wine.target
y_pred = model.predict(df)

print("\nTarget or Actual Values:\n", y_true)
print("Predicted:\n", y_pred)

#สามารถแสดงแผนภาพกราฟจุดของข้อมูล y_true และ y_pred ออกทางจอภาพ
#เลือกใช้ 2 คอลัมน์แรกในการพลอตกราฟ คือ Alcohol และ Malic acid

plt.figure(figsize=(12, 5))

#กราฟแสดงค่าจริง (Actual)
plt.subplot(1, 2, 1)
#ใช้คอลัมน์ 'alcohol' เป็นแกน X และ 'malic_acid' เป็นแกน Y
plt.scatter(df['alcohol'], df['malic_acid'], c=y_true, cmap='prism')
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.title('Actual Targets (Wine)')

#กราฟแสดงค่าที่ทำนายได้ (Predicted)
plt.subplot(1, 2, 2)
plt.scatter(df['alcohol'], df['malic_acid'], c=y_pred, cmap='prism')
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.title('Predicted Clusters (K-Means)')

plt.show()

#สามารถแสดง Confusion Matrix ของ y_true และ y_pred ออกทางจอภาพได้
print("\nConfusion Matrix")
print(metrics.confusion_matrix(y_true, y_pred))

#สามารถประเมินแบบจำลองโดยใช้ Accuracy ได้ โดยแสดงค่าทั้งหมดออกทางจอภาพ
print("\nAccuracy Score")
print(metrics.accuracy_score(y_true, y_pred))
```

2.จงแปลงค่าเลขกลุ่มของผลลัพธ์ y_pred ใน Problem 1 เพื่อให้ได้รับค่าความถูกต้อง (Accuracy) มากสุดเท่าที่จะเป็นไปได้
* สามารถสลับกลุ่มที่เหมาะสมที่สุดของ y_pred  และแสดงค่า y_true และ y_pred ที่สลับใหม่อีกครั้ง
* แสดงแผนภาพกราฟจุดของข้อมูล y_true และ y_pred ใหม่อีกครั้ง
* สามารถแสดง Confusion Matrix ของ y_true และ y_pred ใหม่ ออกทางจอภาพได้
* สามารถประเมินแบบจำลองใหม่โดยใช้ Accuracy ได้ โดยแสดงค่าทั้งหมดออกทางจอภาพ

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets, metrics
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, accuracy_score
from scipy.stats import mode

wine = datasets.load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
y_true = wine.target
# กำหนด random_state เพื่อให้ผลลัพธ์นิ่ง (ถ้าไม่กำหนด เลขกลุ่มอาจเปลี่ยนทุกครั้งที่รัน)
kmeans = KMeans(n_clusters=3, random_state=42)
y_pred = kmeans.fit_predict(df)

#สามารถสลับกลุ่ม y_pred ที่เหมาะสม เพื่อให้ได้รับค่าความถูกต้องมากสุดเท่าที่จะเป็นไปได้
labels = np.zeros_like(y_pred)
for i in range(3):
    mask = (y_pred == i)
    if np.sum(mask) > 0:
        # หาค่าที่ซ้ำกันเยอะที่สุด (Mode) ของ y_true ในกลุ่มนั้นๆ
        labels[mask] = mode(y_true[mask], keepdims=True).mode[0]

y_pred_new = labels

#สามารถแสดงค่า y_true และ y_pred ที่สลับใหม่
print("New Predicted Values")
print(y_pred_new)

#สามารถแสดงแผนภาพกราฟจุดของข้อมูล y_true และ y_pred ใหม่อีกครั้ง
plt.figure(figsize=(12, 5))

# กราฟซ้าย: ข้อมูลจริง (Target)
plt.subplot(1, 2, 1)
plt.scatter(df['alcohol'], df['malic_acid'], c=y_true, cmap='prism')
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.title('Actual Targets (y_true)')

# กราฟขวา: ข้อมูลที่ทำนายและสลับกลุ่มแล้ว (Remapped Prediction)
plt.subplot(1, 2, 2)
plt.scatter(df['alcohol'], df['malic_acid'], c=y_pred_new, cmap='prism')
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.title('New Predicted Values (y_pred_new)')

plt.show()

#สามารถแสดง Confusion Matrix ของ y_true และ y_pred ออกทางจอภาพได้
print("\nConfusion Matrix:")
cm = confusion_matrix(y_true, y_pred_new)
print(cm)

#สามารถประเมินแบบจำลองโดยใช้ Accuracy ได้ โดยแสดงออกทางจอภาพ
acc_score = accuracy_score(y_true, y_pred_new)
print("\nAccuracy Score:")
print(acc_score)

```

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder


#สามารถนำเข้าข้อมูลจาก weather1.csv ได้ และแสดงในรูปแบบตารางได้
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
df = pd.read_csv('/content/drive/MyDrive/Lab271/weather1.csv')

print("แสดงข้อมูลในรูปแบบตาราง")
display(df)
print("\n")

#สามารถแปลงข้อมูลให้อยู่ในรูปแบบ dataset ได้
dataset = df.values.tolist()

print("ข้อมูลในรูปแบบ Dataset")
print(dataset)
print("\n")

#สามารถแปลง dataset ให้อยู่ในรูปแบบ Numpy boolean array
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)

print("Numpy Boolean Array")
print(te_ary)
print("\n")

#สามารถแปลง Numpy boolean array ให้อยู่ใน DataFrame ดังรูปในโจทย์
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

print("DataFrame (DataFrame)")
display(df_encoded)
print("\n")

#สามารถแสดงผล frequent_patterns ได้ โดยใช้ฟังก์ชั่น apriori และกำหนด min_support=0.2
frequent_itemsets = apriori(df_encoded, min_support=0.2, use_colnames=True)

print("Frequent Patterns (Support >= 0.2)")
display(frequent_itemsets)
print("\n")

#สามารถสร้าง และแสดงผล Associate Rules โดยกำหนดตัววัดเป็น 'lift' และกำหนด min_threshold=1.6 ได้
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.6)

print("Association Rules (Lift >= 1.6)")
# แสดงคอลัมน์ที่สำคัญเพื่อให้ดูง่ายขึ้น
cols_to_show = ['antecedents', 'consequents', 'support', 'confidence', 'lift']
display(rules[cols_to_show])
```
